## Slicing arrays

```python
np.arange(3*4*5).reshape((3,4,5))[0:2]
```

```python
np.arange(3*4*5).reshape((3,4,5))[0:2,2:4]
```

```python
np.arange(3*4*5).reshape((3,4,5))[0:2,2:4, 1:3]
```

```python
np.arange(3*4*5).reshape((3,4,5))[0:2,2:4, 1:3].reshape((2,4))
```

So with dimension we have (the naive dimension, the dimensions of each of the naive_dimension objects, etc...)
